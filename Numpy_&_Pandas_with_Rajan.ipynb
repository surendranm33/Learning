# Solving Linear Equations using Numpy
---

Scenario :
<pre>
suppose
Day 1
2 mangoes , 3 oranges cost Rs 34

Day 2
4 Mangoes , 1 Orange Cost Rs 26

we are asked to find price of each Mango & Orange.

Equation :

```
2x + 3y  = 34
4x + 1y  = 26
```
where x represents price of mango and y represents price of orange.

````python
# first we prepare an array containing coefficents

coeff = [[ 2, 3] , [4 ,1] ]

# now we prepare array containing constants

const = [ 34,26 ]
````

``` numpy provides np.linalg.solve() to solve linear equations```

````python
solution = np.linalg.solve(  coeff , const )
````

````python
solution
````

````python
mango_price = solution[0]
orange_price = solution[1]
````

````python
print('Price of Each 🥭 ' , mango_price)
````

````python
print('Price of Each 🍊 ' , orange_price)
````

Let Us Do the SPEED Test Comparision Between Python vs Numpy  !
---
Scenario: we are going to add 1 million numbers using both numpy and python ,one which takes less time to complete will be the winner! ⛳

````python
import time
import numpy as np

SIZE = 1000000

A1 = range(SIZE)
A2 = range(SIZE)

L1 = np.arange( SIZE)
L2 = np.arange(SIZE )

# Python Test
start= time.time()
result =[x+y for x,y in zip(A1,A2)]
stop = time.time()

print('🐍 Python took ' , (stop-start) * 1000 , ' ms ')

# Numpy Test

start = time.time()
result = L1+L2
stop = time.time()

print('💻 Numpy Took '  , (stop-start) * 1000 ,' ms ')
````
