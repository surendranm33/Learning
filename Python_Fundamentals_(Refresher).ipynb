# Python Fundamentals

Objective: ask the user to enter the basic Pay ,
<pre>
if basicPay <=5000 , then calculate HRA as 16% of basicPay , and DA as 14% of basicPay

if basicPay >5000, then Caclulate HRA as 12% of basicPay and DA as 10% of basicPay

find gross Pay i.e. sum of basicPay + HRA + DA

````python

basicPay = int (input('enter basic pay 💴 '))

if basicPay > 5000 :

  HRA= .12 * basicPay
  DA = .1  * basicPay

else:

  HRA = .16 * basicPay
  DA  = .14 * basicPay

# let us find the grossPay

grossPay = basicPay + HRA + DA

# let us use string formatting using F-String to print the salary details

salarySummary =f'''
-------------------------Salary Slip -----------------
BasicPay : {basicPay}
HRA     :  {HRA}
DA      :  {DA}
-------------------
Gross :    {grossPay}
'''

print( salarySummary )
````

Rajan's Encryption Algorithm !
----
( you are provided with an algorithm that you need to translate into python Code )
<pre>

Step 1 - ask the user to enter a plain text to encrypt say 'apple' 🍎

Step 2 - Reverse the given Text i.e.

'apple' becomes 'elppa'

Step 3 - replace Vowels with 'digits' as per the details below:

'a' -> '0'
'e' -> '1'
'i' -> '2'
'o' -> '3'
'u' -> '4'

i.e. 'elppa' becomes '1lpp0'

Step 4: add 'aca' at the end of text

i.e.   '1lpp0' become '1lpp0aca'

print('the encrypted text i.e. 1lpp0aca') END


# Solution / Explanation
---

````python
# text = input('enter text to encrypt ')

text = 'apple'
````

````python
rev_text = text[::-1]

rev_text
````

````python
# we must replace the vowels with digits. 'a'->0 , 'e'->1 ,'i'->2,'o'->3,'u'->4

for letter in 'aeiou' :

  rev_text = rev_text.replace( letter , str( 'aeiou'.index(letter)) )
````
